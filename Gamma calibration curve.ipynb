```python
##import puplot and numpy

import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = (8,8) # Set the plot size

##Generate data

x=[21.9, 158.46, 202, 351.07]
y=[78.2, 553.44, 704.96, 1223.72]
plt.plot(x,y,"o",markersize=5);
plt.title('Magnetic field - Frequency')
plt.ylabel('Frequency (10^8 Hz)')
plt.xlabel('Magnetic Field (mT)')
plt.grid()
plt.show()
```

```python
p, V = np.polyfit(x, y, 1, cov=True);
print ('Gradient '," %6.3E " % p[0],' error'," %6.3e " % np.sqrt(V[0][0]))
print('Intercept'," %6.3e " % p[1],' error'," %6.3e " % np.sqrt(V[1][1]))
```

```python
yy = np.polyval(p, x) #evaluate the fitted function at the x values
plt.plot(x,yy,"-"); #and plot this as a line graph
plt.plot(x,y,"o",markersize=5.0) #Plot the original data for comaparison
plt.grid()
plt.title('Calibration Curve')
plt.ylabel('Centroid Frequency (keV)')
plt.xlabel('ADC')
plt.show()
```
